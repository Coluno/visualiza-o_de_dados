## 1\. Preço do diamante por tipo de corte

```python
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('diamonds')

#criando o grafico
plt.figure(figsize=(10,8))
grafico_diamante = sns.barplot(data = data, x='cut', y = 'price', errorbar= = None, palette = 'pastel', estimator = 'sum', hue = 'cut', legend = False)
grafico_diamante.set(title = 'Preços dos Diamante por tipo de corte', xlabel = 'tipo de Corte', ylabel = 'Preço')
plt.show()
```

**Insight do gráfico 1**: O gráfico de barras empilhadas ilustra o preço total dos diamantes para cada tipo de corte. Podemos observar que os cortes mais elevados, como 'Ideal' e 'Premium', contribuem significativamente para o preço total, indicando que essas categorias contêm diamantes com valores mais altos.

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('diamonds')

plt.figure(figsize=(10,8))
grafico_diamante = sns.barplot(data = data, x = 'cut', y =  'price', palette = 'pastel', hue = 'cut', estimator = 'mean', legend= False)
grafico_diamante.set(title = 'Preços médio dos Diamentes por tipo de corte', xlabel = 'Tipo de corte', ylabel = 'Preço médio')

#loop para adicionar o medio em cada barra

for p in grafico_diamante.patches:
    grafico_diamante.annotate(f'{p.get_height():.0f}', (p.get_x() + p.get_width() / 2. , p.get_height()), ha = 'center', va = 'center', fontsize = 10, color = 'black', xytext = (0,  5), textcoords= 'offset points')

plt.show()
```

**Insight dos gráficos :** O tipo de corte influencia significativamente o preço dos diamantes como foi citado no grafico acima, porém podemos ver que o preço médio de diamentas da categoria 'premium' e 'fair' são maiores. logo percebemos duas coisas
- O segundo gráfico (preço médio) destaca a variação nos preços médios entre os cortes.
- O primeiro gráfico (soma total dos preços) destaca a contribuição relativa de cada tipo de corte para o preço total geral dos diamantes.

Em resumo, o segundo destaca tendências de preços médios, e o primeiro destaca a contribuição total de cada categoria para o preço total

## 2\. Número de passageiros em dezembro por ano

```python
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```python
#data['month'].value_counts()
```

```python
# filtrar a data
dezembro_data = data[data['month'] == 'Dec']

#criando grafico
with sns.axes_style ('whitegrid'):
    grafico = sns.FacetGrid(data = dezembro_data, palette='pastel')
    grafico.map(sns.lineplot,'year','passengers')
    grafico.map(plt.fill_between,'year', 'passengers', alpha = 0.2)
    grafico.set(title = 'Passageiros em dezembro de cada ano', xlabel = 'Ano', ylabel = 'Passageiros')
    grafico.fig.set_size_inches(w=15/2.54, h= 7/2.54)
```

**Insight do gráfico 2**: Podemos observar uma tendência de aumento no número de passageiros de cada ano, indicando um crescimento nas viagens durante esse mês . Este gráfico é valioso para entender a variação anual nos números de passageiros em dezembro, fornecendo insights importantes para o planejamento e análise de tendências sazonais.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```python
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```python
# Filtrar os dados para o intervalo de anos desejado (1955 a 1959)
flights = data.query("1949 <= year <= 1959")

# Criar um gráfico de linha
with sns.axes_style('whitegrid'):
    grafico = sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="pastel")
    grafico.set(title='Passageiros por Mês entre 1955 e 1959', xlabel='Mês', ylabel='Passageiros')
    grafico.get_legend().set_title("Ano")

plt.show()
```

**Insight do gráfico 3**: Podemos observar que há um padrão sazonal consistente, com picos de passageiros em determinados meses. Além disso, ao longo dos anos, parece haver um aumento geral no número de passageiros
